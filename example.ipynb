- In this notebook, we present an example of how to use our module.

- Let's import our functions:

- There are some ways:

```python
from YOUR MODULE import YOUR FUNCTIONS
```
Or

```python
from YOUR MODULE import *
```

Or 

```python
import YOUR MODULE
```

````python
from analysistools.core import read_file, split_text_into_words, split_text_into_sentences, get_capital_words, print_word_counts
````

````python
text = read_file("alice.txt")
type(text)
````

````python
words = split_text_into_words(text)
type(words)
````

````python
sentences = split_text_into_sentences(text)
type(sentences)
````

````python
cap_words = get_capital_words(words)
type(cap_words)
````

````python
print_word_counts(cap_words)
````

````python
help(read_file)
````

- If you use `from YOUR MODULE import *`, it works the same
- If you use `import YOUR MODULE`, then you can call functions as follows:

````python
import analysistools
````

````python
t = analysistools.core.read_file("alice.txt")
type(t)
````

````python
w = analysistools.core.split_text_into_words(t)
type(w)
````

````python
s = analysistools.core.split_text_into_sentences(t)
type(s)
````

````python
d = analysistools.core.get_capital_words(w)
type(d)
````

````python
analysistools.core.print_word_counts(d)
````
